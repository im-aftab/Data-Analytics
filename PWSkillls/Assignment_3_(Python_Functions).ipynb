Q5. What are iterators in Python and how do they differ from iterables?

  - An iterator is an object that represents a stream of data and allows you to access elements sequentially, one at a time.

  Key difference between iterables and iterators are :
    - Purpose:
        - An iterable is a container of items that can be iterated over, while an iterator is an object that performs the iteration by keeping track of the current position.
    - State:
      - Iterables typically store all their elements in memory. Iterators, on the other hand, maintain an internal state (the current position) and only produce one element at a time, making them memory-efficient, especially for large sequences.
    - Creation:
      - Iterables are directly created (e.g., a list literal). Iterators are typically created from iterables using iter().
    - Methods:
      - Iterables primarily implement __iter__() (or __getitem__()). Iterators must implement both __iter__() and __next__().
    - Reusability:
      - Iterables can be iterated over multiple times, as they provide a new iterator each time iter() is called. Iterators are generally "one-shot"; once exhausted (after StopIteration), they cannot be reset and re-used for another iteration without creating a new iterator.

  NOTE :
    - **Iterable:** An object capable of returning its elements one at a time (e.g. list, tuple, string). It uses the __iter__() method.

    - **Iterator:** An object that represents a stream of data; it implements both __iter__() and __next__(). The iterator is obtained from an iterable using iter(). Iterators can only be traversed once.

Q6. Explain the concept of generators in Python and how they are defined.
 - Concept
    - A generator is a special type of iterator in Python that allows you to iterate over a sequence of values lazily—meaning it produces items one at a time, on demand, instead of generating them all at once in memory.

      - Regular functions use return → terminate after returning a value.

      - Generators use yield → pause their execution, remembering their state, and resume when the next value is requested.
    
- Generators are defined like normal functions but use the yield keyword.
  
Key Points:

  - Calling a generator function returns a generator object.
  - next() retrieves the next value until StopIteration is raised.
  - Can be used directly in loops without manually calling next().

Q7. What are the advantages of using generators over regular functions?
  - the advantages of using generators over regular functions are :

    - Memory Efficiency
        - Generators do not store all values in memory — they yield one item at a time.

        Example: Processing a file line-by-line without loading the whole file.

    - Lazy Evaluation
      - They compute values only when needed, which is useful for large datasets or infinite sequences.

    - Statefulness Without Extra Code
        - The generator automatically remembers where it left off without explicit variables or bookkeeping.

    - Pipeline Construction
        - We can chain generators together for data processing pipelines.

Q9. Explain the purpose and usage of the `map()` function in Python.
  - The map() function in Python is a built-in higher-order function used to apply a specified function to each item in an iterable (such as a list, tuple, or set), returning an iterator of the results.

  Usage of map() function :
    - Efficient Transformation:
      - It enables the application of a single function to multiple elements within an iterable without requiring explicit loops, making code more concise and often more performant, especially for large datasets.
    - Data Processing:
      - It is commonly used for transforming data, applying mathematical operations, or performing string manipulations on collections of items.

Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
  - Key Idea
      - map() → transforms every item in a collection.
      - filter() → selects only some items based on a condition.
      - reduce() → combines all items into one result.
  
  Key Difference :
      - **map()**
        - The map() function is used to apply a function to each element of an iterable (like a list, tuple, etc.).
        - It returns a map object (which is an iterator) that can be converted to a list or iterated over.
        - It is typically used for transforming or modifying each element in a collection.

      - filter()
        - The filter() function is used to filter out elements from an iterable based on a condition.
        - The function passed to filter() should return True for elements you want to keep and False for those you want to remove.
        - It returns a filter object (iterator).

    - reduce()
        - Unlike map() and filter(), which are built-in functions, reduce() comes from the functools module and must be imported.
        - reduce() applies a function cumulatively to the items in an iterable, reducing them to a single value.
        - It’s useful for performing aggregations such as sums, products, or combining results.

 Q1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_even_numbers(numbers):
    sum = 0
    for num in numbers:
        if num % 2 == 0:
            sum += num
    return sum

numbers = map(int, input("Enter a list of numbers separated by spaces: ").split())
result = sum_even_numbers(numbers)
print("The sum of all even numbers in the list is:", result)
```

Q2. Create a Python function that accepts a string and returns the reverse of that string.

```python
def reverse_string(string):
    return string[::-1]

string = input("Enter a string: ")
result = reverse_string(string)
print("The reverse of the string is:", result)
```

Q3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

```python
def square_numbers(numbers):
    squares = []
    for num in numbers:
        squares.append(num ** 2)
    return squares

numbers = map(int, input("Enter a list of numbers separated by spaces: ").split())
result = square_numbers(numbers)
print("The list of squares is:", result)
```

Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

number = int(input("Enter a number from 1 to 200: "))
if number < 1 or number > 200:
        print("Invalid input. Please enter a number from 1 to 200.")
        exit()
else:
  if is_prime(number):
    print(number, "is a prime number.")
  else:
    print(number, "is not a prime number.")
```

Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
def fibonacci_generator(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

n = int(input("Enter the number of terms: "))
fibonacci_sequence = fibonacci_generator(n)
print("The Fibonacci sequence up to", n, "terms is:")
for num in fibonacci_sequence:
    print(num)
```

Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_two_generator(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
exponent = int(input("Enter the exponent: "))
power_of_two_sequence = power_of_two_generator(exponent)
print("The powers of 2 up to", exponent, "are:")
for num in power_of_two_sequence:
    print(num)
```

Q7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
# --- Step 1: Create the file with sample content ---
file_name = "rhymes.txt"

# Writing "Twinkle Twinkle Little Star" lines into the file
with open(file_name, "w", encoding="utf-8") as f:
    f.write("Twinkle twinkle little star\n")
    f.write("How I wonder what you are\n")
    f.write("Up above the world so high\n")
    f.write("Like a diamond in the sky\n")

# --- Step 2: Generator function to read the file lazily ---
def read_file_line_by_line(file_path):
    """
    Generator that yields each line of a file as a string without trailing newline.
    """
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            yield line.strip()  # remove \n and extra spaces

# --- Step 3: Use the generator to display lines one at a time ---
print("Reading 'rhymes.txt' line by line:")
for rhyme_line in read_file_line_by_line(file_name):
    print(rhyme_line)
```

Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
init_temp_list = [23, 54, 65, 5, 90, -5, 100]
final_temp_list = list(map(lambda x: (x * 9/5) + 32, init_temp_list))
print(final_temp_list)
```

Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
string = input("Enter a string: ")
vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
result = ''.join(filter(lambda x: x not in vowels, string))
print(result)
```
